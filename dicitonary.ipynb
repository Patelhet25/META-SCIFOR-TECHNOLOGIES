```python
#DICTIONARY

b = {1:"RAM",2:"SHYAM",3:"RAJU"}
print(b)
b.clear()
print(b)
b1 = {1:"RAM",2:"SHYAM",3:"RAJU"}
b1.copy()
print(b1)
x = b1.get(2)
print(x)
b1.items()
print(b1)
b1.update({"4":"HET"})
print(b1)

b1.keys()
print(b1)
b1.values()
print(b1)
```

```python
#SETS

c = {"ram","shyam","raju"}
print(c)
print("raj" in c)
c.discard("ram")
print(c)

x = {"ram","shyam","raju"}
x.remove("ram")
print(x)

a = {"raj","shyam"}
b = {"ram","lakhan"}
a1 = a.union(b)
print(a1)
```

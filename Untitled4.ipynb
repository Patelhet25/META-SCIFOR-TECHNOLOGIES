```python
'''Nidhi challenged her friends to check whether a given number is palindrome or not without using string
manipulation. All her friends were not able to solve this problem. Write a program in python for the
same.
â— Hint: Use a while loop and a temporary variable'''
```

```python
num = int(input("Enter Number:-"))
number = num
zero = 0
while number>0:
  step = number % 10
  zero = (zero * 10) + step
  number = number // 10
if number == zero:
  print("true")
else:
  print("false")
```

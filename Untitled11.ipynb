```python
#calculator

class calculator:
  def __init__(self,number1,number2):
    self.number1 = number1
    self.number2 = number2

class single(calculator):

  def add(self):
    print(self.number1+self.number2)
  def subtraction(self):
    print(self.number1-self.number2)
  def multi(self):
    print(self.number1*self.number2)
  def div(self):
    print(self.number1/self.number2)

number1 = int(input("ENTER NUMBER1:-"))
number2 = int(input("ENTER NUMBER2:-"))

object1 = single(number1,number2)
object1.add()
object2 = single(number1,number2)
object2.subtraction()
object3 = single(number1,number2)
object3.multi()
object4 = single(number1,number2)
object4.div()
```

```python
print("WELCOME TO ATM MACHINE")
balance = 50000

pin = input("ENTER YOUR PIN:-")
if pin == 1234:
  print("YOU HAVE ENTER CORRECT PIN")
else:
  print("YOU HAVE ENTER INCORRECT PIN")

while True:
  d = "deposit"
  w = "Withdraw"
  c = "CheckBalanace"
select = input("Enter Option from Above:-")
if select == d:
  amount = input("Enter Amount:-")
  print("Your total amount in account:-", amount + balance)
elif select == w:
  withdraw = input("Enter Amount to Withdraw:-")
  print("Your total amount in account:-", balance - withdraw)
elif  select == c:
  print(balance)
else:
  print("You have selected wrong Option")


```

```python

print("1:ADD")
print("2:sub")
print("3:multi")
enter = int(input("ENTER NUMBER FROM ABOVE:-"))

number1 = int(input("ENTER NUMBER1:-"))
number2 = int(input("ENTER NUMBER2:-"))

def add():
  add1 = number1 + number2
  print("THE TOTAL OF ADDITION:-",add1)

def sub():
  sub1 = number1 - number2
  print("THE TOTAL OF SUBTRACTION:-",sub1)

def multi():
  multi1 = number1 * number2
  print("THE TOTAL OF MULTIPLICATION:-",multi1)


if enter == 1:
  add()
elif enter == 2:
  sub()
elif enter == 3:
  multi()
else:
  print("WRONG NUMBER")
```

```python
'''
Sahil is wondering whether we can use the same concept of functions in creating simple and
compound interest calculators. He is trying to create it but is unable to return the answers from the functions. Help him in doing the same
'''
```

```python
def simple(a,b,c):
  simple_int = (a*b*c)/100
  return simple_int
def compund(a,b,c):
  compund_int = a*(1+c/100)**b
  return compund_int

a = int(input("ENTER AMOUNT:-"))
b = int(input("ENTER TIME:-"))
c = int(input("ENTER RATE:-"))
print("1:simple interest")
print("2:compund interest")
select = int(input("ENETR NUMBER FROM ABOVE:-"))
if select == 1:
  print("THE SUM OF SIMPLE:-",simple(a,b,c))
elif select == 2:
  print("THE SUM OF COMPUND ",compund(a,b,c))
else:
  print("WRONG NUMBER")
```

```python
# calculator program

class var:

  def __init__(self,x,y):
    self.x = x
    self.y = y

  def add(self):
    return self.x + self.y
  def sub(self):
    return self.x - self.y
  def multi(self):
    return self.x * self.y
  def div(self):
    return self.x / self.y

addition = var(20,20)
addition.add()
```

```python
subtraction = var(200,20)
subtraction.sub()
```

```python
multiplication = var(20,20)
multiplication.multi()
```

```python
division = var(20,20)
division.div()
```

```python
class Question:
    def __init__(self, question, answer):
        self.question = question
        self.answer = answer

class Quiz:
    def __init__(self, questions):
        self.questions = questions
        self.score = 0

    def run(self):
        for question in self.questions:
            self.ask_question(question)

    def ask_question(self, question):
        user_answer = input(question.question)
        if user_answer.lower() == question.answer.lower():
            self.score += 1
            print("Correct!")
        else:
            print("Incorrect. The correct answer is:", question.answer)
        print()

if __name__ == "__main__":
    question_prompts = [
        "Who is the PM of India?\n(a)modi \n(b) rahul\n",
        "What is national animal of India?\n(a) tiger\n(b) lion\n",
        "What is national bird if India?\n(a) sparrow\n(b) peacock\n"
    ]

    questions = [
        Question(question_prompts[0], "a"),
        Question(question_prompts[1], "a"),
        Question(question_prompts[2], "b")
    ]

    quiz = Quiz(questions)
    quiz.run()
```

```python
def prime():
  pre = int(input("ENTER NUMBER:-"))

  for i in range(2, (pre+1)//2):

    if (pre % i) == 0:
      print("IT IS NOT A PRIME NUMBER")
    else:
      print("IT IS A PRIME NUMBER")

prime()

```

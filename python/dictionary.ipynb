```python
#dicitonary

name = {1: 'RAM', 2: 'SHYAM', 3: 'RAJU'}
print(name)

a = name.items()
print(a)

b = name.keys()
print(b)

c = name.pop(2)
print(c)

d = name.popitem()
print(d)

e = name.clear()
print(e)

f = name.values()
print(f)

h = name.update({1: "GANSHYAM",2: "RAJ", 3:"DEEP"})
print(name)

i = name.setdefault(1)
print(i)

j = name.copy()
print(j)

name_1 = {"RAM","SHYAM","RAJU"}
no = {1,2,3}
k = name.fromkeys(name_1,no)
print(k)

l = name.get(3)
print(l)
```

```python
#sets

month = {"JAN","FEB","MARCH","APRIL","MAY","JUNE"}
month_1 = {"AUG","SEP","OCT","JAN"}

co = month.copy()
print(co)

ju = month.add("JULY")
print(month)

di = month.difference(month_1)
print(di)

re = month.difference_update(month_1)
print(month)

inte = month.intersection(month_1)
print(inte)

dis = month_1.discard("OCT")
print(month_1)

po = month.pop()
print(po)

re = month_1.remove("SEP")
print(month_1)

un = month.union(month_1)
print(un)

up = month.update(month_1)
print(month)

sd = month.symmetric_difference(month_1)
print(sd)

jo = month.isdisjoint(month_1)
print(jo)
```

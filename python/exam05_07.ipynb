```python
'''Christiana is a kid who is born on leap year, but she doesn't know which year would be a leap year for her so that
she can celebrate her birthday hence write a python program to help Chirstiana in finding out if the given year is a leap year or not'''
```

```python
year = int(input("Enter Born year:-"))

if year%400 == 0:
  print("YOU BORN IN LEAP YEAR:-", year)
else:
  print("YOU BORN IN NOT LEAP YEAR:-", year)
```

```python
'''Mathew, a 13 year old kid, is good at math but whenever it comes to the even and odd numbers he always faces difficulty to find the even or odd number
greater than 10.
So write a python program to help Mathew to print odd and even numbers in a certain range using a for loop.
'''
```

```python
number = int(input("Enter Number:-"))

for i in number:

  if i % 2 == 0:
    print("THIS IS EVEN NUMBER:-", number)
  else:
    print("THIS IS ODD NUMBER:-", number)
```

```python
'''create a program to the find total number of upper letters,lower letters,digits and the symbols using if elif else, for loop and string operations'''
```

```python
name = input("ENTER THE WORD:-")

L = "abcdefghijklmnopqrstuvwxyz"
U = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
D = 1234567890
S = !@#$%&*

lo = 0
up = 0
dp = 0
sp = 0

for i in name:
  if i in L:
    L+=1
  elif i in U:
    U+=1
  elif i in D:
    D+=1
  elif i in S:
    S+=1
  else:
    print("WRONG")

print("TOTAL UPPER LETTERS:-", up)
print("TOTAL LOWER LETTERS:-", lo)
print("TOTAL DIGIT NUMBER:-", dp)
print("TOTAL SPECIAL LETTERS:-", sp)
```

```python
#Create a program of finding parameter of a triangle and area of any three shapes using class object
```

```python
class traingle:
  def __init__(self,base,height):
    self.base = base
    self.height = height

  def cal(self):
    return 0.5  * self.base * self.height

base = float(input("ENTER THE BASE:-"))
height = float(input("ENTER THE HEIGHT:-"))

Triangle = traingle(base,height)
area = Triangle.cal()
print("THE AREA OF TRIANGLE:-",area)
```

```python
#Create a factorial program using class object
```

```python

class factorial:

  def factoral(self,n):
    number = 1
    for i in range(1, number + 1):
      number = number * i
      return number

n = int(input("ENTER THE NUMBER:-"))


Factorial = factorial()
answer =Factorial.factoral(n)
print("factorial number:-", answer)


```
